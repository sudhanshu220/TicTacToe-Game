# RULES For TIC TAC TOE GAME which is implemented in below code

* Two players required to play this game.
* We will randomly select one first player and another player as second player
* First player will select a marker between 'X' or 'O' and the remaining marker will be assigned to player 2
* Then simuntaneously player1 and player2 will fill the board untill we get our winner or match is draw
* After that we will again ask players that they want to continue the game or not
* If yes then the above processes will be repeated.

```python
from IPython.display import clear_output

def display_board(board):
    """
     Function that can print out a board. Setting up board as a list, where each index 1-9 corresponds 
     with a number on a number pad, so we can get a 3 by 3 board simple representation of board
    """
    clear_output()
    print(board[7]+'|'+board[8]+'|'+board[9])
    print('-----')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('-----')
    print(board[1]+'|'+board[2]+'|'+board[3])
    
def player_input():
    """
    Function that can take in a player input and assign their marker as 'X' or 'O'.
    We have  used a  while loops to continually ask until we get a correct answer which is 'X' or 'O'.
    """
    choice = 'wrong'
    while choice not in ['X','O']:
        #clear_output()
        # Asking player 1 to select between 'X' or 'O' untill they give correct choice
        choice = input('Player 1 please select between "X" or "O": ')
        
        if choice not in ['X','O']:
            print('Not a valid identifier')
    player1 = choice
    if player1 == 'X':
        player2 ='O'
    elif player1 == 'O':
        player2 = 'X'
    else:
        pass
    return (player1,player2) # returning marker of player 1 and 2 as a tuple


def place_marker(board, marker, position):
    """
    Function that takes in the board list object, a marker ('X' or 'O'), and a 
    desired position (number 1-9) and assigns it to the board.

    """
    board[position] = marker
    return board


def win_check(board, mark):
    """
    Function that takes in a board and a mark (X or O) and 
    then checks to see if that mark has won
    Win Condition -- If all the vertical columns are same or all the
    horizontal rows are same or all diagonal elements are same
    
    """
    if ((board[1]==board[2]==board[3]==mark) or (board[4]==board[5]==board[6]==mark)
       or (board[7]==board[8]==board[9]==mark) or (board[7]==board[5]==board[3]==mark)
       or (board[1]==board[5]==board[9]==mark) or (board[1]==board[7]==board[4]==mark)
       or (board[8]==board[5]==board[2]==mark) or (board[3]==board[6]==board[9]==mark)):
        return 'WON!'
    else:
        return 'Not Predicted'
    
def space_check(board, position):
    """
     Function that returns a boolean indicating whether a space on the board is freely available
     True if freely available or else False
    
    """
    if board[position] == ' ':
        return True
    else:
        return False

    
def full_board_check(board):
    """
    Function that checks if the board is full and returns a boolean value. True if full, False otherwise.
    """
    for ind in range(1,len(board)):
        #print(ind)
        if board[ind] == ' ':
            return False
    return True


def player_choice(board):
    """
    Function that asks for a player's next position (as a number 1-9) and 
    then uses the function space_check(board, position) to check if it's a free position. 
    If it is, then return the position for later use.

    """
    position = 'wrong'
    within_range = False
    while position.isdigit() == False or within_range ==False:
        position = input('Enter the position on which you want to fill:- ')
        
        if position.isdigit() == False:
            print('Hey you have entered some wrong datatype')
        else:
            if int(position) in range(1,10):
                within_range = True
            else:
                within_range = False
     
    if space_check(board, int(position)):
        return int(position)
    elif full_board_check(board):
        return 
    else:
        return player_choice(board)
    

def replay():
    """
    Function that asks the player if they want to play again and returns a boolean True
    if they do want to play again. and game continues

    """
    play_again = 'wrong'
    
    while play_again not in ['Y','N']:
        play_again = input('Do you want to paly again?("Y","N"): ')
        
        if play_again not in ['Y','N']:
            print('Please select between "Y","N" ')
    if play_again == 'Y':
        return True
    else:
        return False
    

    
import random

def choose_first():
    """
    Function that uses the random module to randomly decide which player goes first.
    Here we will ask the name of player to make this game more personalize
    and randomly assign any one player as player1 and another player as player2
    and player will play accordingly as per the assigment.
    
    """
    val1 = input('Welcome! Please enter your name ')
    val2 = input('Welcome! Please enter your name ')
    
    player1 = random.choice((val1,val2))
    #print(player1)
    
    if player1 == val1:
        player2 = val2
    else:
        player2 = val1
    #return player1,player2
        
    print(f"Our first player is {player1}")
    print(f"Our second player is {player2}")
    return player1,player2
    #return 'Player1a'
    

print('Welcome to Tic Tac Toe!')

play_again = True
while play_again:
    clear_output()
    #We will set up the agme here
    player1,player2 = choose_first() # Randomly assgin one player as player1 and he will select the marker first
    marker1,marker2=player_input() # player 1 will select the marker and pending marker will be assigned to player2
    board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' '] # setting the board
    #player1,player2 = choose_first() 
    game_on = True
    #clear_output()
    while game_on:
        print(f'{player1} your turn')
        #Player 1 Turn
        palyer1_pos = player_choice(board) # Asking the position to player 1 where he wants to put his marker
        if full_board_check(board) == True and win_check(board, marker1) == 'Not Predicted': #To check the draw condition
            # To check draw condition
            #print ('Draw Match')
            display_board(board)
            print ('Draw Match!!!')
            break
        else:
            place_marker(board, marker1, palyer1_pos) # putting the marrker on the board, position provided by player1
            display_board(board)
            
        if win_check(board, marker1) == 'WON!': # To check the win condition if won then comes out of
            # the loop or else it asks the plyer2 its position where he wants to put marker
            
            game_on = False
            #print('Player1 Wonnnn')
            display_board(board)
            print(f'{player1} WONNN!!!!!')
            break
        
        # Player2's turn.
        print(f'{player2} your turn')    
        palyer2_pos = player_choice(board) # Asking the position to player 2 where he wants to put his marker
        if full_board_check(board) == True and win_check(board, marker2) == 'Not Predicted': 
            # To check draw condition
            #print('Draw Match')
            display_board(board)
            print ('Draw Match !!!!!')
            break
        else:
            place_marker(board, marker2, palyer2_pos) # putting the marrker on the board, position provided by player2
            display_board(board)
            
        if win_check(board, marker2) == 'WON!':
            game_on = False
            #print('Player 2 Wonn')
            display_board(board)
            print(f'{player2} WONNN!!!!!!')
            break
        

    if  replay()==True:
        play_again = True
    else:
        play_again = False
```
